```python
#We will convert the "Class_x" to an integer using for loop
train['target1'] = train['target'].map({'Class_1':0, 'Class_2':1, 'Class_3':2, 'Class_4':3})
train.sample(5)
```

```python
#dropping id from train because we do not need it
train = train.drop(['id'], axis=1)
```

```python
#Checking for outliers
plt.figure(figsize=(18,25))
sns.boxplot(data = train, orient = 'h')
```

```python
#Checking correlation between features
plt.figure(figsize=(18,25))
sns.heatmap(train.corr(method='pearson'), cbar=False, annot=True, fmt='.1f', linewidth=0.1, cmap='coolwarm');
```

```python
#Model Creation
X = train.iloc[:,:-2].values
y = train.iloc[:,-1].values
```

```python
#splitting training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print("Length of X_train is: {X_train}".format(X_train = len(X_train)))
print("Length of X_test is: {X_test}".format(X_test = len(X_test)))
print("Length of y_train is: {y_train}".format(y_train = len(y_train)))
print("Length of y_test is: {y_test}".format(y_test = len(y_test)))
```

```python
#fitting data to model
from xgboost import XGBClassifier, plot_importance
classifier = XGBClassifier(random_state=42, use_label_encoder=True)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
y_score = classifier.predict_proba(X_test)
```
